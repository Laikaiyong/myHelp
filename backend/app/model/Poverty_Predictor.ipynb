```python
import pandas as pd
import matplotlib as mt
import numpy as np
import joblib

file_path = '/content/hies_2019.csv'
data = pd.read_csv(file_path)
data
```

```python
data.drop('area', axis=1, inplace=True)
```

```python
# Identify non-numeric values in 'income_mean'
non_numeric_values = data['income_mean'][~data['income_mean'].astype(str).str.isnumeric()]

# Print unique non-numeric values
print("Non-Numeric Values in 'income_mean':", non_numeric_values.unique())

# Replace non-numeric values with NaN and convert the column to numeric
data['income_mean'] = pd.to_numeric(data['income_mean'], errors='coerce')
```

```python
# Check for null values
null_values = data.isnull().sum()

# Display the count of null values for each column
print("Null Values in Each Column:")
print(null_values)

# Check for empty values (assuming empty strings)
empty_values = (data == '').sum()

# Display the count of empty values for each column
print("\nEmpty Values in Each Column:")
print(empty_values)
```

```python
from pycaret.regression import *

# Assuming your dataset is stored in a DataFrame called 'df'
# Assuming X_test and y_test are defined

# Initialize the PyCaret environment
reg_setup = setup(data, target='poverty_rate', categorical_features=['area_type'])

# Compare different regression models and select the best one
best_model = compare_models()

# Create the final model
final_model = create_model(best_model)

# Tune the hyperparameters of the final model
tuned_model = tune_model(final_model)

# Evaluate the performance of the tuned model
evaluate_model(tuned_model)

# # Save the tuned model for later use
save_model(tuned_model, 'et')


```

```python

model = joblib.load('et.pkl')


# Collect user inputs
area_type = input("Enter area type (state, dun, or district): ")
income_mean = float(input("Enter income mean: "))
expenditure_mean = float(input("Enter expenditure mean: "))
gini = float(input("Enter Gini coefficient: "))

# Create a DataFrame with the user input
user_data = pd.DataFrame({
    'area_type': [area_type],
    'income_mean': [income_mean],
    'expenditure_mean': [expenditure_mean],
    'gini': [gini]
})

# Predict poverty rate
poverty_rate_prediction = model.predict(user_data)[0]

# Display the prediction
print(f"Predicted Poverty Rate: {poverty_rate_prediction}")
```
